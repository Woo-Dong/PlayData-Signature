```python
import pandas as pd
```

```python
df_c = pd.read_csv('covid_count_data.csv')
df_gender = pd.read_csv('gender_data.csv')
df_age = pd.read_csv('age_data.csv')
df_area = pd.read_csv('area_data.csv')
```

```python
df = pd.merge(df_c, df_gender, how='outer',on='date')
```

```python
df = pd.merge(df, df_age, how='outer',on='date')
df
```

```python
df = pd.merge(df, df_area, how='outer',on='date')
```
